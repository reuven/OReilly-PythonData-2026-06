# What's going on here?

1. We created a new `Series` value, by invoking `pd.Series` on a list.
2. Our series needs to have values that are all of the same type! In this case, they're all integers.
3. The series can have any number of values, but once defined, that number cannot change.
4. Behind the scenes, the series hides a NumPy array of integers. (Pandas is a shell around the low-level NumPy library.)
5. When we print our series, you see that we are given both the index (numbers 0-9) and the values.
6. We can see the dtype of the series, which is `int64`.

In Pandas, though, when we add two arrays together, it adds them according to the index -- so whatever is at index 0 in both is added together, etc.

The two arrays that you're adding together both need to be of the same size.

# Broadcasting

We've seen that if I perform an arithmethic operation on two series, I get a new series back, the result of performing that operation on every pair of values, based on their index.

```python
highs - lows
```

```python
highs * lows
```

```python
# What if I try the same thing, but not with two series, but with one series and one scalar number?
# answer: broadasting. The operation is "broadcast," performing it on every element of the array,
# and returning a new array.

highs + 5
```

```python
lows - 15   # go to each element of lows, and subtract 15
```

```python
highs * 3
```

# Who cares?

This is a fundamental skill and mindset to use when working with Pandas, because it sits at the center of everything we do in retrieving data.

We will eventually use broadcasting with `True`/`False` values, and thus retrieve particular elements from our series. 

# `.loc` vs. `.iloc`

Right now, there is no difference between the index and the position. However, we will soon see that these are different -- that you can have an index that doesn't contain 0, 1, 2, 3, etc.

In such cases (and often others), you'll want to retrieve data based on the index or the position:

- `.loc` retrieves based on the index, and does a *lot*. `.loc`, if you give it a slice, will go through the final value.
- `.iloc` retrieves based on the position, and works like normal Python, up to and *not* including the final value.

```python
# what if, instead of passing a list of indexes, I were to pass a list of booleans?
# It turns out that this is fine!
# we get back all of the elements of the series where there was a True value.
# the False values are ignored.

s.loc[ [True, False, False, False, False, True, True] ]
```

```python
s.loc[ [False, True, False, False, True, False, False] ]
```

We can agree that while this is interesting, it doesn't seem that practical. Who is going to create a series of boolean values? 

We can return to our friend, broadcasting, for some help. Until now, we've thought of broadcasting as something we do with numbers. What if we were to use, instead, comparison operators that return `True` and `False`, such as `==` or `<` or `>`?

```python
s < 50   # is each element of s < 50?
```

```python
# 1. calculate s < 50, get a boolean series back (broadcast)
# 2. apply that boolean series to s
# 3. get back the elements of s where the boolean was True
# 4. in the end, we get back elements of s that are < 50

s.loc[ s < 50 ]
```

```python
s.loc[ s < s.mean() ]   # get only those values that are < mean
```

```python
# what about this?

s.loc[ s % 2 == 1 ]
```

```python
# let's try this again

# 1. s.loc[ s < 50 ]
# whatever is done inside of [] is evaluated first

s < 50
```

```python
s.loc[s<50] # we get the elements of s that are < 50
```

```python
# 2. s.loc[ s < s.mean() ] 

# 1. evaluate s.mean()
# 2. compare each value in s with s.mean()
# 3. this gives a boolean series
# 4. apply the boolean series to s

s < s.mean() 
```

```python
s.loc[ s < s.mean() ] 
```

```python
# s.loc[ s % 2 == 1 ]

# 1. calculate s%2
# 2. compare s%2 with 1

s.loc[s%2 == 1]
```

# Dtypes

What's the difference between a Python list and a Pandas series? The series has, at its core, a NumPy array.

Better: How are lists different from arrays?

- Lists can change their size over time, whereas arrays cannot.
- All elements in a list should be of the same type, but in arrays they *must* be of the same type.

All of the elements in a Pandas series must be of the same type. That type is known as the "dtype". These dtypes reflect the C storage that is used for our series. 

```python
s = Series([10, 20, 30])
s
```

```python
# we can retrieve the dtype with the .dtype attribute

s.dtype
```

When we define a new series, we can specify what the dtype should be, using the "dtype" keyword. What are some possibilities?

- Integers
    - `int64` -- default if you define the series with only ints
    - `int32`
    - `int16`
    - `int8`
- Unsigned integers
    - `uint64`
    - `uint32`
    - `uint16`
    - `uint8`
- Floats
    - `float64` -- default if you define the series with only floats
    - `float32`
    - `float16`


```python
s = Series([10, 20, 30], dtype='float64')
s
```

```python
s = Series([10, 20, 30], dtype='int32')
s
```

What's the point of using a different dtype? Why not just use `int64`, the default?

If you're dealing with millions or even billions of data points, then you have to worry about how much memory you'll use. Pandas can only work with data that is in memory.

- If you have 1b data points, and you use `int64`, that's 256 GB.
- If you have 1b data points, and you use `int16`, that's 64 GB.

So we don't want to get too big. What about too small?

The problem is that you can have a dtype that's too small for some of your numeric values.

```python
s = Series([20, 80, 100], dtype='int8')
```

```python
s + 10
```

```python
s + 100  # this is not good, though!
```

When choosing a dtype, you need to think as big as possible (what if you have billions of data points) and also as small as possible (what is the smallest dtype that won't corrupt the data?)

```python
s
```

```python
# I need to do some big calculations with s, and I worry that I'll have problems

# I can  use the .astype method to get a new series back from the existing one.
```

```python
s.astype('int64')
```

```python
# what if I decide that I want to set the dtype just by assignment
# don't do it!

s
```

```python
s.dtype = 'int16'
```
