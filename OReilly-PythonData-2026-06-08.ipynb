# Jupyter notebook

Many Python programmers like to use a REPL -- read, eval, print loop. This is highly interactive, and allows us to explore our code and our data with almost no overhead. 

Jupyter is a Web-based REPL. You run it on your computer, it pops up inside of your browser, and is a combination system of browser + server. You type into Jupyter, and it then runs any Python code you have. It gives you the illusion of running the code in the browser, even though it's really running on the server.
