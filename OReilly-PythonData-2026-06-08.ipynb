# Jupyter notebook

Many Python programmers like to use a REPL -- read, eval, print loop. This is highly interactive, and allows us to explore our code and our data with almost no overhead. 

Jupyter is a Web-based REPL. You run it on your computer, it pops up inside of your browser, and is a combination system of browser + server. You type into Jupyter, and it then runs any Python code you have. It gives you the illusion of running the code in the browser, even though it's really running on the server.

# A 5-minute intro to Jupyter

We type into *cells* in Jupyter. Each cell can contain code (Python) or documentation (Markdown). Each cell has two "modes":

- Edit mode -- when I type into the cell, the text appears. I'm in edit mode right now! You can enter edit mode by pressing ENTER or by clicking inside of cell.
- Command mode -- when I type a character/letter, that is interpreted as a command by Jupyter. You can edit command mode by pressing ESC or by clicking to the left of a cell.

What commands are available in command mode?

- `c` -- copy the current cell
- `x` -- cut the current cell
- `v` -- paste the most recently cut/copied cell
- `a` -- add a new cell *above* the current one
- `b` -- add a new cell *below* the current one
- `m` -- make a cell Markdown documentation
- `y` -- make a cell Python code

When you're done in a cell, whether it's writing documentation or code, you can press shift+ENTER together, and the cell will "run."

# Data science

A big, poorly defined term! I think about it as being several overlapping things:

- Data analytics, which is making sense of data that we've already gathered. (That's what we'll be doing here.)
- Data engineering, which is about getting data from one place to another.
- Modeling and machine learning, which are about making predictions based on existing data.

All of these together are data science.

Many companies now have people working in all three of these kinds of departments.

# Python is the #1 language for data

This makes sense, because Python executes super fast, and uses very little memory, right?

No. Not at all.

Python is actually very fast and very efficient when used for data. That's because of NumPy, a module written mostly in C and a little in Python. NumPy gives you super-fast, super-efficient storage of numbers and other data, with a thin Python layer on top.

NumPy is a bit low level. If you're looking to read data in from CSV files, clean the data, and produce visualizations based on your analysis, NumPy is fine... but probably a bit annoying to work with.

That's why we're going to learn about Pandas, which is a higher-level library that uses (for the time being) NumPy.

Pandas gives you the best of both worlds -- the ease of Python, and the speed of NumPy/C. You can read from files, clean the data, analyze the data, produce plots, and output files.
