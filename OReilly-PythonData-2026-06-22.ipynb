# Sorting

What does it mean to *sort* our data?

- We want to order it, typically from smallest to largest *or* from largest to smallest
- When it comes to data in Pandas, we have many more options than just those

```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('taxi.csv',                          # load the 10k taxi records
                 index_col=['tpep_pickup_datetime'])  # when creating the data frame, make the pickup datetime the index

df
```

```python
# I want to find the shortest trip distances
# I can grab the trip_distance column and run min on it

df['trip_distance'].min()
```

```python
# what about the 5 shortest or the 10 shortest?
# to most easily do this, we will invoke the sort_values method on the series

df['trip_distance'].sort_values()  # this returns a new series, based on the original, sorted in ascending order
```

```python
df['trip_distance'].sort_values().head(5)  # get the 5 smallest values
```

```python
df['trip_distance'].sort_values().head(15)
```

```python
# what about the largest values?

df['trip_distance'].sort_values().tail(15)  # keep the sort in ascending order, and get the largest (ending) values
```

```python
# another option: sort in *descending* order

df['trip_distance'].sort_values(ascending=False).head(5) # take the first 5 elements/value, after sorting in descending order
```

# Even more sorting!

It's natural to understand what `sort_values` does. But there's another way to sort a data frame (or series) in Pandas, that is to sort the *index*. For that, we have a separate `sort_index` method. The result is a series in which the index is sorted, and the values just come along for the ride.

```python
df['trip_distance'].sort_index()  # we'll get back a new series in which the trips are from earliest to latest.
```

# When to use each one?

- If you want the largest/smallest values, then `sort_values` is more appropriate and useful.
- If you want to get the values sorted by datetime, or user ID, or address, then you can make that column into the index, and run `sort_index`.

# What about sorting data frames?

Generally speaking, every method that's available on a series is also available on a data frame.

- `sort_index` is easier to understand on a data frame; it works just like on a series, but multiple columns come along for the ride.
- `sort_values` is more complex, because you have many values that you could sort by. You have to specify to Pandas which column you want to sort by.

```python
df.sort_index()
```

```python
df.sort_values('trip_distance')
```

```python
# What if I want the dropoff times for the 5 longest trips?

(
    df
    .sort_values('trip_distance')  # sort by the trip distance, in ascending order
    ['tpep_dropoff_datetime']      # grab the one column that's of interest to us
    .tail(5)                       # get the 5 final values, corresponding to the largest trip_distance
)
```

# Can we sort by more than one column?

We saw that when we invoke `sort_values`, we can pass a column name. We've seen, though, that wherever you can pass a single column name (a string) in Pandas, you can also pass a list of column names (a list of strings). We can do that here, too!



```python
(
    df
    .sort_values(['trip_distance', 'total_amount'])  # if two trips have the same trip_distance, then total_amount will break the tie
)
```

# Two more sorting methods

These methods are shortcuts for sorting + grabbing the largest and smallest values:

- `nlargest` -- `df.nlargest(columns='trip_distance', n=10)` -- this returns the rows with the 10 largest `trip_distance` values
- `nsmallest`

```python
df.nlargest(columns='trip_distance', n=10)
```

# Grouping

Let's say that I have the taxi data loaded up, and I want to know, for rides with 1 passenger, what was the mean `trip_distance`?

```python
df = pd.read_csv('taxi.csv')

(
    df
    .loc[ pd.col('passenger_count') == 1 ]   #  replaces this: .loc[ df['passenger_count'] == 1 ]
    ['trip_distance']
    .mean()
)
```

```python
# now do it for trips with 2 passengers

(
    df
    .loc[ pd.col('passenger_count') == 2 ] 
    ['trip_distance']
    .mean()
)
```

```python
# now do it for trips with 3 passengers

(
    df
    .loc[ pd.col('passenger_count') == 3 ] 
    ['trip_distance']
    .mean()
)
```

# Grouping is precisely for this problem:

- We have a categorical column with a number of distinct values
- We also have a column with numeric values
- We want to perform an aggregation/calculation for each distinct categorical value, running on the numeric values

For example: I want the mean of `trip_distance` for each distinct value of `passenger_count`

```python
#             categorical       numeric         aggregation method
df.groupby('passenger_count')['trip_distance'].mean()
```

# Examples of where/how to use grouping

- For each country, what was the total number of sales?
- For each salesperson, what was the mean number of sales this year?
- For each product, what is the max price?

You'll always need an aggregation method, typically: `mean`, `median`, `max`, `min`, `sum`, `count`

# Exercise: Olympic grouping

1. For each team, what was the mean athlete height? Which 5 teams were tallest, on average?
2. For each team, what was the number of medals they accumulated? Note that the `count` method ignores `NaN`.


```python
olympics_df = pd.read_csv('olympic_athlete_events.zip')

```

```python
# For each team, what was the mean athlete height? Which 5 teams were tallest, on average?

# - categorical - team, I want one result for each unique team name
# - numeric - Height
# - aggregation method will be mean

olympics_df.groupby('Team')['Height'].mean()  # this returns a series -- index is team names, values are mean height for each team
```

```python
# on a series, we can invoke nlargest, but there are columns, so we just pass a number

(
    olympics_df
    .groupby('Team')['Height'].mean()
    .nlargest(5)
)
```

```python
# For each team, what was the number of medals they accumulated? Note that the count method ignores NaN.

# categorical - Team
# to-be-calculated - Medal
# aggregation method - count

(
    olympics_df
    .groupby('Team')['Medal'].count()
    .nlargest(10)
)
```

# Next up

- More grouping
- Pivot tables

If you can say, "for every _____, calculate the ____," then that's a `groupby`.

We can be more sophisticated in our grouping, too.

```python
# we want to get the mean trip_distance for every distinct value of passenger_count

#             categorical       numeric         agg method
df.groupby('passenger_count')['trip_distance'].mean()
```

```python
# what if I want to get the mean trip_distance and *also* the mean total_amount?

df.groupby('passenger_count')['total_amount'].mean()
```

```python
# we can provide a list of columns inside of the [] -- so yes, we'll have [[ ]]

#            categorical        list of numeric columns          agg method
df.groupby('passenger_count')[['trip_distance', 'total_amount']].mean()
```

# Examples of how we might use this

- For each country, what were the median population and income?
- For each city, what is are the mean family size and sq meters/feet per home?

```python
# What if we want to calculate more than one aggregation method?
# then ... we just put a list there!

#            categorical       numeric            agg methods -- two of them!
df.groupby('passenger_count')['trip_distance'].agg(['mean', 'std'])
```

```python
# we can also provide a list of categorical columns
# in this case, I get back a series *but* the index has two parts!
# this is known as a multi-index

df.groupby(['passenger_count', 'VendorID'])['trip_distance'].mean()
```

# Bottom line:

- we can have 1 or more categoricals
- we can have 1 or more data/numeric columns
- we can have 1 or more aggregation methods

Can we have multiples of more than one of these? YES, but then it gets very complex.

# Exercise: Olympic grouping

1. Load the Olympic data into a data frame.
2. For each sport, find the mean height and weight.
3. For each combination of year and season, find the mean height
4. For each sport, find the mean + std height.

```python
olympics_df
```

```python
# For each sport, find the mean height and weight.

# categorical - sport
# numeric/data - Height + Weight
# agg function: mean

olympics_df.groupby('Sport')[['Height', 'Weight']].mean()
```

```python
# For each combination of year and season, find the mean height

# categorical: Year + Season
# numeric/data: Height
# agg function: mean

olympics_df.groupby(['Year', 'Season'])['Height'].mean()
```

```python
# For each sport, find the mean + std height.

# categorical: Sport
# numeric/data: Height
# aggregation function: mean + std

olympics_df.groupby('Sport')['Height'].agg(['mean', 'std'])
```

```python
# what if we go completely bananas, and use multiple items in each portion of the groupby?

# categorical: Year + Season
# numeric/data: Height + Weight
# aggregation methods: mean + std

olympics_df.groupby(['Year', 'Season'])[['Height', 'Weight']].agg(['mean', 'std'])
```
