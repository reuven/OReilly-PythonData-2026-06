# Method chaining

Pandas methods generally assume that you don't want to change the data frame, but that you instead want to get a new data frame back, reflecting the changes you made.

For example, in the case of `rename`, we got back a new data frame that was different from `df`, in that a column was renamed. But it didn't change the column permanently.

If you want to change it permanently, you can (should) assign back to the variable

    df = df.rename(.....)

You can (but SHOULD NOT) pass the `inplace=True` keyword argument to a lot of methods. This does change the data frame in place, and returns `None`.

Why not do this?

1. You won't be able to do method chaining, meaning that you won't be able to invoke a bunch of transformations on your data frame.
2. It's often assumed that using `inplace=True` is faster, or uses less memory, than getting a new data frame back. THIS IS NOT THE CASE.


```python
# if I want to remove a column, I can use df.drop, passing a list to the "columns" keyword argument

df.drop(columns=['tax'])   # this returns a new data frame, without 'tax'
```

```python
df = df.drop(columns=['tax'])  # the right way to do it
```

# Retrieving with `.loc` -- the 2-argument version

We've used `.loc` to retrieve an element from a series and a row from a data frame.

It turns out  that we can also invoke `.loc` with two arguments:

- The first is a "row selector"
    - one index value
    - a list of index values ("fancy indexing")
    - boolean index to retrieve only those element where it's `True`
- The second is a "column selector"
    - one column name
    - a list of column names

```python
df
```

```python
df.loc[
    123   # row selector
    ,
    'department'   # column selector
    ]
```

```python
df.loc[
    [123, 789]     # row selector
    ,
    'department'   # column selector
    ]
```

```python
df.loc[
    [123, 789]     # row selector
    ,
    ['department', 'product']   # column selector
    ]
```

# Method chaining

In Python, you can typically have one statement/expression per line. However, if you open `()`, Python sees everything inside of them as being on one line.

This lends itself to writing long, complex Pandas queries across multiple lines, where each row "inherits" the output from the previous one.

```python
(
    df
    .set_index('t') 
    .loc[
        ['hello', 'today'],   # row selector -- these two rows
         ['x', 'z']          # column selector -- these two columns
         ]
    .sum()
)
```

```python
# if you do set the index permanently (by assigning to a variable), you can also
# invoke reset_index, and turn the index back into a regular column

(
    df
    .set_index('t') 
    .loc[['hello', 'today']]
    .reset_index()
    [['t', 'y']]
)
```

# Remember `nan`?

`NaN` (or `nan`), is "not a number." This is what data sets use to indicate that they're missing values, or the values are unknown.

On a series:

- `dropna` returns a new series without the orignal's `NaN` values
- `fillna` returns a new series where `NaN` is replaced by whatever you give it

These methods also exist on data frames!

- If you invoke `dropna`, then any *row* containing `NaN` is removed
- If you invoke `fillna`, then any `NaN` value is replaced. If you pass a single value as the replacement, then it's used everywhere. But if you pass a series, then you can actually replace different values in different columns.

```python
df
```

```python
df = df.drop(columns=['t'])
df
```

```python
df.loc['b', 'y'] = np.nan
df.loc['a', 'z'] = np.nan
df.loc['a', 'x'] = np.nan
df.loc['d', 'y'] = np.nan

df
```

```python
# let's check the dtypes of our columns

df.dtypes  
```

```python
# what happens if I invoke df.dropna()? Every row with even one NaN is removed

df.dropna()  # this returns a new data frame, based on df...
```

```python
df
```

```python
df.fillna(999)
```

```python
df.mean()  # remember, NaN isn't included in this calculation
```

# Boolean indexes

You might remember from last week that Pandas uses a weird kind of syntax, a "boolean index" (aka a "mask index") to retrieve selected values from a series.

The idea is:

- We invoke, via broadcasting, an operator on a series
- We get back a series of `True` and `False` values
- We apply that boolean series, via `.loc`, to our series of values
- Only those that correspond to `True` are kept

```python
s.loc[[True, False, True]]  # very simple version
```

```python
s.loc[ s >= 20 ]  # get all values of s that are >= 20
```

```python
# we can do the same thing with a data frame!
# our row selector (i.e., the first argument to .loc) can be a boolean series!

df
```

```python
df['price'] > 30
```

```python
# traditional way to select rows in Pandas

df.loc[   df['price'] > 30  ]
```

```python
# as of Pandas 3.0 (from January), we have a new, better syntax

df.loc[   pd.col('price') > 30  ]
```

When I started to work with data, I was sure that there were great, well-thought-out file formats. That is now starting to be the case, but it usually isn't.

The most common file format is CSV (comma-separated values), which is basically a table on disk:

- Every line in the file is a row in the data frame
- Every line broken up (usually with commas) into fields

The problem is that everything in a CSV file is text. So Pandas has to figure out which columns are integers and floats, and it often can't do that on its own.
